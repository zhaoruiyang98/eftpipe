```python
import numpy as np
import matplotlib.pyplot as plt
from eftpipe.reader import read_pkl
```

```python
def plot_window(s, Q0, Q2, Q4, Q6, Q8):
    plt.semilogx(s, Q0, label=R"$\ell=0$")
    plt.semilogx(s, Q2, label=R"$\ell=2$")
    plt.semilogx(s, Q4, label=R"$\ell=4$")
    plt.semilogx(s, Q6, label=R"$\ell=6$")
    plt.semilogx(s, Q8, label=R"$\ell=8$")
    plt.legend(frameon=False)
    plt.xlabel(R"$s\,[\mathrm{Mpc}\,h^{-1}]$")
    plt.ylabel(R"$\mathcal{W}_\ell(s)$")
    plt.grid()
    plt.xlim(left=1)
    plt.ylim(-0.1, 0.8)
    plt.show()
```

```python
s, Q0, Q2, Q4, Q6, Q8 = np.loadtxt("window/Arnaud_ELG_NGC.txt", unpack=True)
plot_window(s, Q0, Q2, Q4, Q6, Q8)
s, Q0, Q2, Q4, Q6, Q8 = np.loadtxt("Arnaud_ELG_data/Window_ELGPk_NGC_0.7z1.1.txt", unpack=True)
plot_window(s, Q0, Q2, Q4, Q6, Q8)
```

```python
s, Q0, Q2, Q4, Q6, Q8 = np.loadtxt("window/Arnaud_ELG_SGC.txt", unpack=True)
plot_window(s, Q0, Q2, Q4, Q6, Q8)
s, Q0, Q2, Q4, Q6, Q8 = np.loadtxt("Arnaud_ELG_data/Window_ELGPk_SGC_0.7z1.1.txt", unpack=True)
plot_window(s, Q0, Q2, Q4, Q6, Q8)
```

```python
data = read_pkl("DR16/Arnaud_pix64_ELG_SGC.txt")
err = np.sqrt(np.loadtxt("cov/pix64_cov_SGC_E024.txt").diagonal().reshape((3, -1)))
k = data.k
estyle = dict(capsize=2)
_ = plt.figure(figsize=(6.5, 4))
plt.errorbar(k, k * data.P0, k * err[0], fmt=".", c="k", label=R"$\ell=0$", **estyle)
plt.errorbar(k, k * data.P2, k * err[1], fmt=".", c="b", label=R"$\ell=2$", **estyle)
plt.errorbar(k, k * data.P4, k * err[2], fmt=".", c="g", label=R"$\ell=4$", **estyle)
plt.title(R"ELG SGC sample, $z_\mathrm{eff}=0.853$")
plt.legend(frameon=False, loc="lower center", ncol=3)
plt.xlim(0.027, 0.254)
plt.ylim(-499, 549)
plt.minorticks_on()
plt.tick_params(which="both", direction="in", top=True, right=True)
plt.grid(color="grey", alpha=0.8, ls=(0, (5, 5)))
plt.xlabel(R"$k\,[h\,\mathrm{Mpc}^{-1}]$")
plt.ylabel(R"$kP_\ell(k)\,[\mathrm{Mpc}/h]^2$")
plt.show()

data = read_pkl("DR16/Arnaud_pix64_ELG_NGC.txt")
err = np.sqrt(np.loadtxt("cov/pix64_cov_NGC_E024.txt").diagonal().reshape((3, -1)))
k = data.k
estyle = dict(capsize=2)
_ = plt.figure(figsize=(6.5, 4))
plt.errorbar(k, k * data.P0, k * err[0], fmt=".", c="k", label=R"$\ell=0$", **estyle)
plt.errorbar(k, k * data.P2, k * err[1], fmt=".", c="b", label=R"$\ell=2$", **estyle)
plt.errorbar(k, k * data.P4, k * err[2], fmt=".", c="g", label=R"$\ell=4$", **estyle)
plt.title(R"ELG NGC sample, $z_\mathrm{eff}=0.86$")
plt.legend(frameon=False, loc="lower center", ncol=3)
plt.xlim(0.027, 0.254)
plt.ylim(-499, 549)
plt.minorticks_on()
plt.tick_params(which="both", direction="in", top=True, right=True)
plt.grid(color="grey", alpha=0.8, ls=(0, (5, 5)))
plt.xlabel(R"$k\,[h\,\mathrm{Mpc}^{-1}]$")
plt.ylabel(R"$kP_\ell(k)\,[\mathrm{Mpc}/h]^2$")
plt.show()
```
